### Лемматизация

```python

data_pur=[] #где будем создавать список из уникальных значений
dt = data['purpose'].unique()  #Уникальные значения в столбце purpose
data_lem=[] #итоговый список всех лемматизированных слов

for text in dt: #создаем список из уникальных значений
    data_pur.append(text)
#print(data_pur)    
from pymystem3 import Mystem
m = Mystem() 
for i in data_pur:
    for word in i.split(): #т.к. на вход поступают словосочетания, надо проверить каждое слово
        #i - поступающее на вход словосочетание, i.split() - словосочетание развитое на список слов в нем, 
        #чтобы проверить каждое слово из словосочетания
        lemmas = ' '.join(m.lemmatize(word)) 
        data_lem.append(lemmas)
        #print(lemmas)

#print(data_lem)  
from collections import Counter
print(Counter(data_lem))

print()
print(data['purpose'].unique())
spisok_lem_itog = ['недвижимость', 'автомобиль', 'образование', 'жилье', 'свадьба'] #список слов, 
#которые встречаются в каждой строке столбца 'purpose'

def lemma_str(text): #создаем правило для лемматизации каждой строки
    lemmas = m.lemmatize(text)
    return lemmas

data['lemma_vse'] = data['purpose'].apply(lemma_str) #новый столбец с леммами по столбцу 'purpose'




print('__________________________________________________')   
display(data)
print('__________________________________________________')        
#Проверяю уникальные значения в столбцах, чтобы понять как их объединить в группы
print()
print('Уникальные значения в столбце education_reg')
print(data['education_reg'].value_counts())
print()
print('Уникальные значения в столбце family_status')
print(data['family_status'].value_counts())
print()
print('Уникальные значения в столбце income_type')
print(data['income_type'].value_counts())
print()
print('Уникальные значения в столбце children')
print(data['children'].value_counts())
print()
```

### Категоризация данных

```python
#Уникальные значения в столбце family_status
#женат / замужем          12380
#гражданский брак          4177
#Не женат / не замужем     2813
#в разводе                 1195
#вдовец / вдова             960
#Произвожу замену, чтобы осталось 3 основные группы в "family_status"
data['family_status'] = data['family_status'].replace('в разводе', 'Не женат / не замужем')
data['family_status'] = data['family_status'].replace('вдовец / вдова', 'Не женат / не замужем')
print('Уникальные значения в столбце family_status')
print(data['family_status'].value_counts())
print('______________________________________________________________________')
#Уникальные значения в столбце family_status_id
# 0    12380 соответствует женат / замужем
# 1     4177 соответствует гражданский брак
# 4     2813 соответствует Не женат / не замужем 
# 3     1195 соответствует в разводе
# 2      960 соответствует вдовец / вдова
# в разводе = Не женат / не замужем , значит 3 заменим на 4
# вдовец / вдова = Не женат / не замужем , значит 2 заменим на 4
#Произвожу замену, чтобы осталось 3 основные группы в "family_status_id"

# 0 - женат / замужем
# 1 - гражданский брак
# 4 - Не женат / не замужем и в разводе и вдовец / вдова

data['family_status_id'] = data['family_status_id'].replace(3, 4)
data['family_status_id'] = data['family_status_id'].replace(2, 4)

#Произвожу замену, т.к. компаньон и предприниматель по смыслу одно и тоже,
#а студент и в декрете относится к безработным
data['income_type'] = data['income_type'].replace('компаньон', 'предприниматель')
data['income_type'] = data['income_type'].replace('студент', 'безработный')
data['income_type'] = data['income_type'].replace('в декрете', 'безработный')


#категоризируем цель кредита объединяя недвижимость и жилье,
# таким образом остается 4 цели - недвижимость, авто, образование, свадьба
def proverka(text):
    if ('недвижимость' in text) or ('жилье' in text):
        return 'недвижимость'
    if 'автомобиль' in text:
        return 'автомобиль'
    if 'образование' in text:
        return 'образование'
    return 'свадьба'

data['purpose_group'] = data['lemma_vse'].apply(proverka) #новый столбец, в котором
# будет только 4 вида цели кредита
#print(data.head())    

#проверим уникальные значения в столбце
print('Уникальные значения в столбце purpose_group')
print(data['purpose_group'].value_counts())
print('______________________________________________________________________')

#образование можно категоризировать на - (среднее, начальное), 
#(неоконченное высшее, высшее, ученая степень)
def obrazovanie(text):
    if ('среднее' in text) or ('начальное' in text):
        return 'до среднего'
    return 'высшее (+ неоконченное)'

data['education_group'] = data['education_reg'].apply(obrazovanie) #новый столбец 

print('Уникальные значения в столбце education_group')
print(data['education_group'].value_counts())
print()
print('______________________________________________________________________')

#есть смысл категоризировать по детям на 0, 1,2 и многодетные - 3,4,5
def deti(text):
    if text == 0:
        return 0
    if text == 1:
        return 1
    if text == 2:
        return 2
    return '3 и более'
data['children_group'] = data['children'].apply(deti) #новый столбец 
print('Уникальные значения в столбце children_group')
print(data['children_group'].value_counts())
print()
print('______________________________________________________________________')
#есть смысл безработных к предпринимателям:), т.к. предприниматели часто предоставляют справки
#о доходе с собственной подписью, не всегда банк требует справку 2 НДФЛ
#ну и предпринимательство более нестабильная ниша в плане дохода в отличии
#от госслужащих и пенисионеров, а также безработных всего 4, что не повлияет на статистику
def prof(text):
    if 'сотрудник' in text:
        return 'сотрудник'
    if 'пенсионер' in text:
        return 'пенсионер'
    if 'госслужащий' in text:
        return 'госслужащий'
    return 'предприниматель'

data['income_type_group'] = data['income_type'].apply(prof) #новый столбец 
print('Уникальные значения в столбце income_type_group')
print(data['income_type_group'].value_counts())
print()
print('______________________________________________________________________')




#из предыдущих вычислений максимальный доход 2 млн.руб. а миним. - 20 тыс.руб.,  
#и доход свыше 500тыс.руб всего 222 чел, свыше 1 млн.руб - 25 чел.
#медиана = 145017

def dengi(sum):
    if sum < 30000:
        return 'do 29 тысяч'
    elif sum < 50000:
        return '30 - 49 тысяч'
    elif sum < 80000:
        return '50 - 79 тысяч'
    elif sum < 100000:
        return '80 - 99 тысяч'
    elif sum < 150000:
        return '100 - 149 тысяч'
    elif sum < 200000:
        return '150 - 199 тысяч'
    elif sum < 300000:
        return '200 - 299 тысяч'
    elif sum < 400000:
        return '300 - 399 тысяч'
    else:
        return 'выше 400 тысяч рублей'

data['itogo_doxod'] = data['total_income'].apply(dengi) #новый столбец 
print('Уникальные значения в столбце itogo_doxod')
print(data['itogo_doxod'].value_counts())
print()
print('______________________________________________________________________')

data = data[['children_group', 'dob_years', 'education_group', 'family_status', 'family_status_id',
             'income_type_group', 'debt', 'itogo_doxod', 'purpose_group']] #таблица, которая
#получилась в результате группировки данных, старые данные заменила на сгруппированные

# в 'family_status_id' следующие категории:
# 0 - женат / замужем
# 1 - гражданский брак
# 4 - Не женат / не замужем и в разводе и вдовец / вдова
def deti_brak(row):
    deti = row[0]
    brak = row[4]
    if deti == 0 and brak == 0:
        return 'женат / замужем, детей нет'
    if deti == 1 and brak == 0:
        return 'женат / замужем, ребенок - 1'
    if deti == 2 and brak == 0:
        return 'женат / замужем, ребенок - 2'
    if deti == 0 and brak == 1:
        return 'гражданский брак, детей нет'
    if deti == 1 and brak == 1:
        return 'гражданский брак, ребенок - 1'
    if deti == 2 and brak == 1:
        return 'гражданский брак, ребенок - 2'
    if deti == 0 and brak == 4:
        return 'Не женат / не замужем, детей нет'
    if deti == 1 and brak == 4:
        return 'Не женат / не замужем, ребенок - 1'
    if deti == 2 and brak == 4:
        return 'Не женат / не замужем, ребенок - 2'
    return 'все остальные'

data['deti_brak'] = data.apply(deti_brak, axis=1) #новый столбец, функцию ко всей строке 
print('Уникальные значения в столбце deti_brak')
print(data['deti_brak'].value_counts())
print()
print('______________________________________________________________________')





print('ИТОГОВАЯ ТАБЛИЦА, НА ОСНОВАНИИ КОТОРОЙ БУДЕТ ПРОИЗВОДИТЬСЯ АНАЛИЗ')
#Заменим старые столбцы (на основании которых категорезировала) на новые - 'children_group',
#'education_group', 'purpose_group', а 'gender', family_status_id, 'days_employed' - уберу из таблицы,
#т.к. они для анализа не пригодятся, 'family_status' - будет достаточно
data = data[['children_group', 'dob_years', 'education_group', 'family_status', 'deti_brak',
             'income_type_group', 'debt', 'itogo_doxod', 'purpose_group']]
display(data)   
data.info()

```

## Ответы на вопросы

- Есть ли зависимость между наличием детей и возвратом кредита в срок?

```python
def vivodi(df, columns_group, columns_stroki):
    data_new=df.pivot_table(index=[columns_group], columns='debt',
                                values=columns_stroki, aggfunc='count')
    data_new['vsego']=data_new[0]+data_new[1]
    data_new['% должников']=((data_new[1] * 100) / data_new['vsego']).astype('int')
    return data_new

vivodi(data, 'children_group', 'purpose_group')
```

Но заемщиков с 3 и более детей - 1,7% от всех заемщиков, поэтому их можно было отнести группе 2 детей.
